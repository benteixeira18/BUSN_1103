```python
# loading dataset
pre_break = pd.read_csv('pre_break.csv')
pre_break.head(3)
```

```python
# removing unneccesary rows and columns: rows
pre_break = pre_break.drop(labels = [0,27],axis=0)
pre_break.head(3)
```

```python
# dropping columns
pre_break = pre_break.drop(pre_break.columns[[1,2,3,4,14,15,16,17,18,19,28,29,30,34,35,37,38,39,40,
                                              41,42,43,44,45,76,77,78,79,80,81,88,89,90,91,92,93]],axis=1)
pre_break.head(3)
```

```python
# finding average course grade
pre_break.sum()
```

```python
# turning grades series into list
grades = pre_break['Current Score'].tolist()
```

```python
# turning list into dataframe to convert into floats
grades = pd.DataFrame({'grades':[float(grade) for grade in grades]})
```

```python
sum_grades = grades['grades'].sum()
avg_grade = sum_grades/len(pre_break)
round(avg_grade,2) # average grade: 85.45. B
```
